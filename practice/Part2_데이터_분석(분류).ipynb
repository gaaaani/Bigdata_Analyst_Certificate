## DecisionTree Classifier(의사결정 나무 분류 모델)


### 1.필요 패키지 import

```python
# package import
import numpy as np
import pandas as pd
import sklearn
```

```python
# decisionTree - classifier
from sklearn.tree import DecisionTreeClassifier

# data split
from sklearn.model_selection import train_test_split
```

### 2.Data 불러오기 + EDA
- iris data

```python
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/iris.csv')
```

```python
df.info()
```

```python
df.describe()
```

### 3.데이터 전처리
- 이상치, 결측치 대체 등
- 데이터 인코딩, 단위 환산, 정규화, 파생변수 생성 등

```python
df['species'].unique()
```

```python
# 명목형 변수에 대하여 Label Encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['species_le'] = le.fit_transform(df['species'])

# pandas의 replace를 이용해서도 가능함
# df['species'].replace({'setosa' : 0, 'versicolor' : 1, 'virginica':2},  inplace = True)
```

### 4.분석 데이터셋 준비
- train/test 분리
- 모델 적합

```python
# X는 독립변수, y는 종속변수
X = df[['sepal_length','sepal_width','petal_length','petal_width']]
y = df['species_le'] ## 라벨링 한 값 사용

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state =11)
```

```python
# 분리된 데이터셋 크기 확인
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# 모델 적합
dt = DecisionTreeClassifier(random_state=11)
dt.fit(X_train, y_train) ## 분석 수행
```

```python
# 학습이 완료된 dt에 대해 분류(예측) 수행
pred = dt.predict(X_test)
print(pred)
```

### 5.성능 평가
- 정확도
- confusion matrix
- report : 오차행렬에 기반한 평가지표



```python
# 분류 결과와 실제 분류 결과 비교
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, pred)
print(acc)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, pred)
# 실제와 같게 예측 됨 -> 28개
```

```python
# 평가지표 표
from sklearn.metrics import classification_report
rpt = classification_report(y_test, pred)
print(rpt)
```

✔ Precision (정밀도)

예측한 것 중에 진짜 맞은 비율
→ 예측한 것 중 얼마나 정확했나  

✔ Recall (재현율)

실제 맞아야 할 것 중에서 얼마나 맞췄나
→ 놓치지 않고 잘 찾았나

✔ F1-score (조화평균)

Precision과 Recall을 모두 고려한 점수
→ 불균형 데이터일 때 중요한 지표  
정밀도와 재현율의 조화평균, 둘 가 높아야 F1-score도 높게 나옴

✔ Support

해당 클래스가 Test Set에서 몇 개 있었는지
→ 단순 개수

## Parc1 - Titanic 데이터셋

### 1.필요 패키지 import

```python
import numpy as np
import pandas as pd
import sklearn

# DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier

# train_test split
from sklearn.model_selection import train_test_split
```

### 2.Data 불러오기 + EDA

```python
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/titanic.csv')
```

```python
df.head()
```

```python
df.info() ## 결측치, 타입 등 확인
```

```python
df.describe()
```

### 3.데이터 전처리
- 나이 -> 평균 대체
- cabin -> 결측치 너무 많아서 분석에서 제외
- embarked -> 최빈값 대체
- 성별, embarked : LabelEncoder이용 -> 순서 생김

```python

# 나이 평균 대체
age_mean = df['Age'].mean()
df['Age'].fillna(age_mean, inplace=True)
print(df['Age'].isna().sum())
```

```python
# cabin 컬럼 제외
# embarked 최빈값 대체
embarked_mode = df['Embarked'].mode()[0]
print(embarked_mode)
df['Embarked'].fillna(embarked_mode, inplace = True)
print(df['Embarked'].isna().sum())
```

```python
# sex, embarkde encoding
df['Sex'].unique()

from sklearn.preprocessing import LabelEncoder
df['Sex'] =  LabelEncoder().fit_transform(df['Sex'])
df['Embarked'] = LabelEncoder().fit_transform(df['Embarked'])
```

```python
# 파생변수 생성
# SibSp(형재 또는 배우자 수), Parch(부모 또는 자녀 수)의 값을 더해서 FamilySize컬럼 생성
df['FamilySize'] = df['SibSp']+df['Parch']
df.head()
```

### 4.분석 데이터셋 준비 및 분석 수행

```python
# data split
X = df[['Pclass','Sex','Age','Fare', 'Embarked', 'FamilySize']]
y = df['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=11)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
dt = DecisionTreeClassifier(random_state=11)
dt.fit(X_train, y_train) ## 학습 수행
```

```python
# test set으로 예측 수행
pred = dt.predict(X_test)
print(pred)
```

### 5.성능평가
- 0 : 사망, 1: 생존

```python
# confusion matrix 통해서 모델 성능 확인
from sklearn.metrics import confusion_matrix

mat = confusion_matrix(y_test, pred)
print(mat)
## 정밀도(precision) = (98/(98+18)) : 예측이 참인 것 중에서 실제 참인 비율
## recall = (98/(98+20)) : 실제 참인 것 중에서 참으로 예측한 비율
```

```python
# report 확인
from sklearn.metrics import classification_report
rpt = classification_report(y_test, pred)
print(rpt)
```

```python
# accuracy 확인
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, pred))
```

## SVM - titanic

### 1.필요 패키지 import

```python
# 기본 패키지 import 생략
# svm 패키지만 import
from sklearn import svm
```

### 2.데이터 불러오기 + EDA
- EDA 생략

```python
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/titanic.csv')
```

### 3.데이터 전처리
- Sex, Embarkde 컬럼 원핫인코딩

```python
# age 평균값 대체
age_mean = df['Age'].mean()
df['Age'].fillna(age_mean, inplace=True)

# embarked 최빈값 대체
embarked_mode = df['Embarked'].mode()[0]
df['Embarked'].fillna(embarked_mode, inplace = True)

# 파생변수 생성
df['FamilySize'] = df['SibSp'] + df['Parch']
```

```python
# 원 핫 인코딩
oh_sex = pd.get_dummies(df['Sex']).astype(int)
df = pd.concat([df, oh_sex], axis=1)
oh_embarked = pd.get_dummies(df['Embarked']).astype(int)
df = pd.concat([df, oh_embarked], axis=1)
```

```python
df.head()
```

### 4.분석 데이터셋 준비 및 수행

```python
X = df[['Pclass', 'Age', 'Fare', 'FamilySize', 'female', 'male','C','Q','S']]
y = df['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
```

```python
# 모델 적합을 위한 svm.SVC() 사용 - rbf 커널 사용
svm = svm.SVC(kernel='rbf')
svm.fit(X_train, y_train)
```

```python
pred = svm.predict(X_test)
print(pred)
```

### 5.성능평가

```python
from sklearn.metrics import classification_report
rpt = classification_report(y_test, pred)
print(rpt)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, pred)
```

### svm 커널 조정
- rbf
- linear C=0.1, gamma = 0.1
- rbf, C=0.1, gamma = 0.1

```python
from sklearn import svm
sv1 = svm.SVC(kernel='linear', C=0.1, gamma = 0.1)
sv1.fit(X_train, y_train)

#예측
pred = sv1.predict(X_test)
# print(pred)

# 성능평가
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pred)) ## 앞선 모델 보다 성능이 좋음
```
