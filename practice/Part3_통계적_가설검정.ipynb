## 1.í‘œë³¸í‰ê·  ê²€ì •
- ëª¨ë¶„ì‚° ì• : z-test

```python
# ì˜ˆì œ ë°ì´í„° ìƒì„±
data = np.array([23, 25, 28, 30, 26, 27, 29, 32, 31, 28])

# ëª¨ì§‘ë‹¨ì˜ í‰ê· ê³¼ ë¶„ì‚°
pop_m = 28
pop_v = 4

# í‘œë³¸ì˜ í‰ê· ê³¼ ë¶„ì‚°
sam_m = np.mean(data)
sam_v = np.var(data)

# ëª¨í‰ê· ê³¼ í‘œë³¸í‰ê· ì´ ê°™ì€ì§€ í™•ì¸í•´ë³´ì

# ìœ ì˜ìˆ˜ì¤€
alpha = 0.05
n = len(data)

# ê²€ì •í†µê³„ëŸ‰
test_statistic = (sam_m - pop_m)/((np.sqrt(pop_v))/np.sqrt(n))

# critical_value
cv = norm.ppf(1-alpha/2)

# p-value
p_val = 2 * (1-norm.cdf(abs(test_statistic)))
```

```python
# ê²°ê³¼ ì¶œë ¥
print("sample mean: " ,sam_m)
print("test_statistic: ", test_statistic)
print("critical value: ", cv)
print("p-value: ", p_val)

if(abs(test_statistic) > cv):
  print("H0 ê¸°ê°í•  ê·¼ê±°ê°€ ì¶©ë¶„í•˜ë‹¤. ëª¨í‰ê· ê³¼ í‘œë³¸ í‰ê· ì€ ë‹¤ë¥´ë‹¤.")
else:
    print("H0 ê¸°ê°í•  ê·¼ê±°ê°€ ì¶©ë¶„í•˜ì§€ ëª»í•˜ë‹¤.")
```

- ëª¨ë¶„ì‚° ëª¨ë¦„ : t-test

```python
from scipy import stats
```

```python
# ì˜ˆì œ ë°ì´í„° ìƒì„±
data = np.array([23, 25, 28, 30, 26, 27, 29, 32, 31, 28])

# ìœ ì˜ìˆ˜ì¤€
alpha = 0.05

# t-test
t_statistic, p_val = stats.ttest_1samp(data, popmean = 5, alternative = "two-sided")

# confidence interval
ci = stats.t.interval(1-alpha, len(data) - 1, loc = np.mean(data), scale = stats.sem(data))
```

```python
# ê²°ê³¼ ì¶œë ¥
print("t-statisitc: ", t_statistic)
print("p-value: ", p_val)
print("confidence interval: ", ci)

if p_val < alpha:
  print("reject H0")
else:
  print("can not reject H0")
```

## 5.ëª¨ë¶„ì‚° ê²€ì • - ì¹´ì´ì œê³± ê²€ì •
- ë‹¨ì¼í‘œë³¸
- stats.chi2.cdf(x, df)
    - x : ì¹´ì´ì œê³±ì˜ ëˆ„ì  í™•ë¥ ì„ êµ¬í•˜ë ¤ëŠ” í†µê³„ëŸ‰
    - df : n-1

```python
from scipy.stats  import chi2

# data
df = np.array([14.2, 15.1, 14.8, 15.4, 14.9, 15.2, 14.6, 15.0])

# H0:ëª¨ë¶„ì‚°ì€ 1.0ê³¼ ê°™ë‹¤ VS H1: ë‹¤ë¥´ë‹¤.
pop_var = 1.0

# ê²€ì •í†µê³„ëŸ‰ = (n-1)*í‘œë³¸ë¶„ì‚°/ëª¨ë¶„ì‚°
sam_var = np.var(df) #í‘œë³¸ë¶„ì‚°
n = len(df)
chi2_stat = (n-1) * sam_var / pop_var

# df
df = n-1

# p-value
p_val = 1-chi2.cdf(chi2_stat, df)
```

```python
# ê²°ê³¼ í™•ì¸
print("chi-square stat : ", chi2_stat)
print("p_val : ", p_val)

if p_val < 0.05:
  print("reject h0, ëª¨ë¶„ì‚°ì€ 1.0ê³¼ ë‹¤ë¥´ë‹¤")
else:
  print("can not rejcet h0")
```

## 6.ëª¨ë¶„ì‚°ë¹„ ê²€ì • - F ê²€ì •
- ë¶„ì‚°ë¹„ ê²€ì •
- stat.f.cdf(x, dfn, dfd)
  - x : Fë¶„í¬ ëˆ„ì  í™•ë¥ ì„ êµ¬í•˜ëŠ” Fí†µê³„ëŸ‰
  - dfn : ë¶„ì ììœ ë„, ì²« ë²ˆì§¸ ê·¸ë£¹ì˜ ììœ ë„
  - dfd : ë¶„ëª¨ ììœ ë„, ë‘ ë²ˆì§¸ ê·¸ë£¹ì˜ ììœ ë„
- F-í†µê³„ëŸ‰ : ë¶„ì‚°ë¹„, 1ë³´ë‹¤ í¬ë©´ ì²« ë²ˆì§¸ ê·¸ë£¹ì˜ ë¶„ì‚°ì´ ë” í¼

```python
# pakage
from scipy import stats
np.random.seed(123)
sam1 = np.random.normal(loc=5, scale=2, size=50)
sam2 = np.random.normal(loc=5, scale=3, size=50)
print(sam1)
print(sam2)
```

```python
# ë‘ í‘œë³¸ì˜ ë¶„ì‚° ê³„ì‚°
sam1_var = np.var(sam1, ddof = 1) #í‘œë³¸ë¶„ì‚°
sam2_var = np.var(sam2, ddof = 1)
print("sam1_var : ", sam1_var)
print("sam2_var : ", sam2_var)
```

```python
# F-test
f_statisitc = sam1_var / sam2_var
df1 = len(sam1)-1
df2 = len(sam2)-1

p_val = 1-stats.f.cdf(f_statisitc, df1, df2)

print("F-í†µê³„ëŸ‰", f_statisitc)
```

```python
# ê²°ê³¼
if p_val < 0.05:
  print("reject h0, ë‘ í‘œë³¸ì˜ ëª¨ë¶„ì‚° ë‹¤ë¦„")
else:
  print("can not reject h0, ë‘ í‘œë³¸ì˜ ëª¨ë¶„ì‚° ë‹¤ë¥´ë‹¤ê³  í•  ìˆ˜ ì—†ìŒ")
```

## 7.ë…ë¦½ì„± ê²€ì • - ì¹´ì´ì œê³± ê²€ì •

- stats.chi2_contigency(observed, correction=True, lambda_=None)
    - observed : í–‰ì´ ë²”ì£¼1ì„ ë‚˜íƒ€ë‚´ê³  ì—´ì´ ë²”ì£¼ 2ë¥¼ ë‚˜íƒ€ë‚´ì•¼ í•¨
    - correction : T/F ììœ ë„ê°€ 1ì¸ ê²€ì •ì˜ ë³´ì •ì„ ì‚¬ìš©í• ì§€ ì—¬ë¶€ ì§€ì •, ê¸°ë³¸ê°’ True
    - lambda_: ëë‹¤ê°’ ë³´ì •

```python
# ë‘ ë²”ì£¼í˜• ë³€ìˆ˜ê°„ì˜ ê´€ê³„ íŒŒì•…
# ìŠ¤í¬ì¸  ë³€ìˆ˜ì™€ ê¸°ë…í’ˆ êµ¬ë§¤ ì‚¬ì´ëŠ”
# h0 : ê´€ê³„ ì—†ìŒ VS h1 : ê´€ê³„ ìˆìŒ

from scipy.stats import chi2_contingency

# ì£¼ì–´ì§„ ë°ì´í„°ë¥¼ ë°°ì—´ë¡œ ì…ë ¥
data = np.array([
    [50, 75],
    [125, 175],
    [90, 30],
    [45, 10]
])
chi2, p, dof, expected = chi2_contingency(data)

print("ì¹´ì´ì œê³± í†µê³„ëŸ‰:", chi2)
print("p-value:", p)
print("ììœ ë„:", dof)
print("ê¸°ëŒ€ë„ìˆ˜:\n", expected)
if p < 0.05:
  print("reject h0, ì—°ê´€ì„± ìˆìŒ")
```
