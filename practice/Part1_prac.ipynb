## 04_Apply, Map


```python
import pandas as pd
import numpy as np
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv")
df.head()
```

```python
print(df.shape)
```

```python
#Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라 Unknown : N
# Less than $40K : a
# $40K - $60K : b
# $60K - $80K : c
# $80K - $120K : d
# $120K +’ : e

dic = {
    'Unknown'        : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K'    : 'b',
    '$60K - $80K'    : 'c',
    '$80K - $120K'   : 'd',
    '$120K +'        : 'e'
}
df['newIncome'] = df['Income_Category'].map(lambda x: dic[x])
df.head()
```

```python
def changeCategory(x):
    if x =='Unknown':
        return 'N'
    elif x =='Less than $40K':
        return 'a'
    elif x =='$40K - $60K':
        return 'b'
    elif x =='$60K - $80K':
        return 'c'
    elif x =='$80K - $120K':
        return 'd'
    elif x =='$120K +' :
        return 'e'

df['newIncome']  =df.Income_Category.apply(changeCategory)
df.head()
```

```python
# Customer_Age의 값을 이용하여 나이 구간을 AgeState 컬럼으로 정의하라. (0~9 : 0 , 10~19 :10 , 20~29 :20 … 각 구간의 빈도수를 출력하라
def age(x):
  if x < 10:
    return 0
  elif x < 20:
    return 10
  elif x < 30:
    return 20
  elif x < 40:
    return 30
  elif x < 50:
    return 40
  elif x < 60:
    return 50
  elif x < 70:
    return 60
  elif x < 80:
    return 70

df['AgeState'] = df['Customer_Age'].apply(age)
df['AgeState'].value_counts()
```

```python
# Education_Level의 값중 Graduate단어가 포함되는 값은 1 그렇지 않은 경우에는 0으로 변경하여 newEduLevel 컬럼을 정의하고 빈도수를 출력하라
df['newEduLevel'] = df['Education_Level'].str.contains("Graduate").astype(int)
df['newEduLevel'].value_counts()
```

```python
# Credit_Limit 컬럼값이 4500 이상인 경우 1 그외의 경우에는 모두 0으로 하는 newLimit 정의하라. newLimit 각 값들의 빈도수를 출력하라
def limit(x):
  if x >= 4500:
    return 1
  else:
    return 0

df['newLimit'] = df['Credit_Limit'].apply(limit)
df['newLimit'].value_counts()
```

```python
# Marital_Status 컬럼값이 Married 이고 Card_Category 컬럼의 값이 Platinum인 경우 1 그외의 경우에는 모두 0으로 하는 newState컬럼을 정의하라. newState의 각 값들의 빈도수를 출력하라
def check(x):
  if x.Marital_Status =='Married' and x.Card_Category =='Platinum':
      return 1
  else:
    return 0
df['newState'] = df.apply(check,axis=1)
df['newState'].value_counts()
```

```python
# Gender 컬럼값 M인 경우 male F인 경우 female로 값을 변경하여 Gender 컬럼에 새롭게 정의하라. 각 value의 빈도를 출력하라
df['Gender'] = df['Gender'].map({"M":"male", "F":"female"})
df['Gender'].value_counts()
```

```python
# Yr_Mo_Dy에 존재하는 년도의 유일값을 모두 출력하라
# 여기서는 dt가 핵심이다.
df['Yr_Mo_Dy'].dt.year.unique()
```

```python
# Yr_Mo_Dy에 년도가 2061년 이상의 경우에는 모두 잘못된 데이터이다.
# 해당경우의 값은 100을 빼서 새롭게 날짜를 Yr_Mo_Dy 컬럼에 정의하라
def fix_year(x):
  import datetime as dt
  year = x.year - 100 if x.year >= 2061 else x.year
  return pd.to_datetime(dt.date(year, x.month, x.day))
df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(fix_year)
df.head()
```

```python
# 년도별 각컬럼의 평균값을 구하여라
df_ym = df.groupby(df['Yr_Mo_Dy'].dt.year).mean()
df_ym.head()
```

```python
df.info()
```

```python
# 주말이면 1, 평일이면 0
def wc(x):
  if x == 5 or x == 6:
    return 1
  else:
    return 0

df['WeekCheck'] = df.weekday.apply(wc)
df.head()
```

```python
# 년도, 일자 상관없이 모든 컬럼의 각 달의 평균을 구하여라
df.groupby(df.Yr_Mo_Dy.dt.month).mean().head()
```

```python
# 모든 결측치는 컬럼기준 직전의 값으로 대체하고 첫번째 행에 결측치가 있을경우 뒤에있는 값으로 대채하라
df = df.fillna(method='ffill').fillna(method="bfill")
df.isnull().sum()
```

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv")
df.head(10)
```

```python
# 년-월-일:시 컬럼을 datetime형태로 변경
import datetime as dt
def change_date(x):
  hour = x.split(":")[1]
  date = x.split(":")[0]

  if hour == '24':
    hour = '00:00:00'

    fin_date = pd.to_datetime(date+" "+hour)+dt.timedelta(days=1)
  else:
    hour = hour+":00:00"
    fin_date = pd.to_datetime(date+" "+hour)
  return fin_date

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)
df['(년-월-일:시)'].info()

# datetime64형태로 변경
# df['Yr_Mo_Dy'] = pd.to_datetime(df['Yr_Mo_Dy'])
# df['Yr_Mo_Dy'].info()
```

```python
#df3과 df4 데이터를 하나의 데이터 프레임으로 합쳐라. 모든 컬럼을 포함하고, 결측치는 0으로 대체한다
pd.concat([df3, df4]).fillna(0)
```
